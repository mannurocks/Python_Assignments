# 1. What is NumPy, and why is it widely used in Python?
NumPy is a Python library used for numerical and scientific computing. It provides support for arrays, matrices, and many mathematical functions to operate on these arrays efficiently. It is widely used due to its performance, ease of use, and extensive functionality for linear algebra, Fourier transforms, and random number generation.

# 2. How does broadcasting work in NumPy?
Broadcasting is a technique used by NumPy to perform arithmetic operations on arrays of different shapes. It automatically expands the smaller array to match the shape of the larger array without copying data, allowing efficient vectorized operations.

# 3. What is a Pandas DataFrame?
A DataFrame is a two-dimensional, tabular data structure provided by the Pandas library. It is similar to an Excel sheet or SQL table and allows for easy data manipulation and analysis using labeled rows and columns.

# 4. Explain the use of the groupby() method in Pandas.
The groupby() method in Pandas is used to group data based on one or more columns and apply aggregate functions like sum, mean, count, etc., to each group. It simplifies data summarization and analysis.

# 5. Why is Seaborn preferred for statistical visualizations?
Seaborn is preferred because it is built on top of Matplotlib and provides a high-level interface for creating attractive and informative statistical graphics. It supports complex visualizations like heatmaps, box plots, and violin plots with minimal code.

# 6. What are the differences between NumPy arrays and Python lists?
- NumPy arrays are more efficient and faster for numerical operations.
- Arrays support broadcasting and vectorized operations.
- Arrays are homogeneous, whereas lists can hold different data types.
- NumPy arrays consume less memory than lists.

# 7. What is a heatmap, and when should it be used?
A heatmap is a graphical representation of data where individual values are represented using color gradients. It is used to visualize correlations, patterns, and concentrations in data, especially in matrices and correlation tables.

# 8. What does the term “vectorized operation” mean in NumPy?
Vectorized operations are operations that are applied element-wise to arrays without using loops. They are implemented at a lower level in C for better performance, making code cleaner and much faster.

# 9. How does Matplotlib differ from Plotly?
Matplotlib is a static, 2D plotting library for Python. Plotly, on the other hand, supports interactive plots, 3D plotting, and dashboards. Plotly is better for web applications, while Matplotlib is widely used for static reporting and publications.

# 10. What is the significance of hierarchical indexing in Pandas?
Hierarchical indexing (MultiIndex) allows users to work with higher-dimensional data in a 2D DataFrame. It enables more complex groupings, reshaping, and advanced data manipulation like pivoting and unstacking.

# 11. What is the role of Seaborn’s pairplot() function?
The pairplot() function in Seaborn creates a matrix of scatter plots to show pairwise relationships between multiple variables in a dataset. It is useful for exploring correlations and distributions at a glance.

# 12. What is the purpose of the describe() function in Pandas?
The describe() function provides summary statistics of numerical columns in a DataFrame, including count, mean, standard deviation, min, max, and quartile values. It is used for quick data exploration.

# 13. Why is handling missing data important in Pandas?
Handling missing data ensures that the dataset is clean and reliable for analysis. Missing values can distort results, cause errors in computations, and affect the performance of machine learning models.

# 14. What are the benefits of using Plotly for data visualization?
- Interactive plots with zoom, pan, and tooltips.
- Easy integration with web apps and dashboards.
- Support for 3D plotting and animations.
- Beautiful, publication-ready charts with minimal code.

# 15. How does NumPy handle multidimensional arrays?
NumPy provides the ndarray object that can represent arrays of any dimension. It allows slicing, reshaping, and mathematical operations along any axis. NumPy handles these arrays efficiently using strides and internal memory layout.

# 16. What is the role of Bokeh in data visualization?
Bokeh is a Python library for interactive and real-time data visualization in web browsers. It allows building complex dashboards with JavaScript-like interactivity using only Python code.

# 17. Explain the difference between apply() and map() in Pandas.
- map() is used to apply a function to each element in a Series.
- apply() is more flexible and can be used on both Series and DataFrames to apply a function along an axis (rows or columns).

# 18. What are some advanced features of NumPy?
- Broadcasting and vectorized operations
- Linear algebra and matrix decomposition
- Fast Fourier transforms
- Random number generation
- Advanced indexing and slicing
- Structured arrays and masked arrays

# 19. How does Pandas simplify time series analysis?
Pandas provides built-in support for datetime objects, frequency-based indexing, resampling, shifting, and rolling operations, which makes time series data manipulation and analysis easier.

# 20. What is the role of a pivot table in Pandas?
A pivot table in Pandas summarizes data by grouping and aggregating it over one or more categorical variables. It helps reshape data for reporting and analysis.

# 21. Why is NumPy’s array slicing faster than Python’s list slicing?
NumPy uses contiguous memory and internal pointers for slicing, making it faster. Python lists are collections of objects and require more overhead during slicing due to reference-based access.

# 22. What are some common use cases for Seaborn?
- Visualizing statistical distributions
- Plotting correlation matrices
- Creating box plots, violin plots, and bar plots
- Pairwise comparisons using pairplot()
- Exploring relationships using regplot() and lmplot()


```python
# Practical Questions

# 1. Create a 2D NumPy array and calculate the sum of each row
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = arr.sum(axis=1)
print("Row-wise sum:", row_sums)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

data = {'Marks': [80, 75, 90, 85]}
df = pd.DataFrame(data)
mean_marks = df['Marks'].mean()
print("Mean Marks:", mean_marks)
```

```python
# 3. Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

plt.scatter(x, y)
plt.title("Simple Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 4. Calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [2, 3, 4, 5]
})

correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 5. Generate a bar plot using Plotly
import plotly.express as px

data = {'Fruits': ['Apples', 'Bananas', 'Cherries'], 'Quantity': [10, 20, 15]}
df = pd.DataFrame(data)

fig = px.bar(df, x='Fruits', y='Quantity', title='Fruit Quantity Bar Plot')
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column
df = pd.DataFrame({'Marks': [50, 60, 70, 80]})
df['Grade'] = ['Pass' if mark >= 60 else 'Fail' for mark in df['Marks']]
print(df)
```

```python
# 7. Element-wise multiplication of two NumPy arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print("Element-wise multiplication:", result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib
x = [1, 2, 3, 4]
y1 = [10, 15, 20, 25]
y2 = [20, 25, 30, 35]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title("Multiple Lines Plot")
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
df = pd.DataFrame({'Score': [50, 60, 70, 80, 90]})
filtered_df = df[df['Score'] > 70]
print(filtered_df)
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution
data = [20, 22, 23, 20, 21, 25, 26, 23, 24]
sns.histplot(data, kde=True)
plt.title("Distribution Histogram")
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 3]])
result = np.dot(A, B)
print("Matrix multiplication result:\n", result)
```
